```python
image_array = np.array(image_grayed)

# dimensions of the image
height, width = image_array.shape[:2] # extract only the height and width, ignoring channel

# calculate the coordinates for the center 150x150 pixel section
top = (height - 150) // 2
left = (width - 150) // 2
bottom = top + 150
right = left + 150

# extract the middle 150x150 pixel section using array slicing
cropped_image = image_array[top:bottom, left:right]

# display the cropped image using matplotlib
plt.imshow(cropped_image, cmap='gray')
plt.axis('off')  # Hide the axes
plt.title("Cropped 150x150 Center Section")
plt.show()
```

## Exercise - 3
### Image Compression and Decompression using PCA.

1. Load and Prepare Data:
*   Fetch an image of you choice. {If colour convert to grayscale}
*   Center the dataset - Standaridze the Data.
*   Calculate the covaraince matrix of the Standaridze data.

```python
image_array = np.array(image_grayed)
data = image_array.copy()

plt.imshow(data, cmap='gray')
plt.axis("off")
plt.show()

print(f"\nShape of Data: {data.shape}")
```

```python
# subtract the mean from data, centered data has zero mean
mean = np.mean(data, axis = 0)
centered_data = data - mean

print(f"Shape of Centered Data: {centered_data.shape}")
```

```python
# covariance matrix
cov_matrix = np.cov(centered_data, rowvar = False) # by default, rows are considered as feature so it is kept as false here (we are working with column)

print(f"Shape of Covariance Matrix: {cov_matrix.shape}")

print(f"\nCovariance Matrix: \n{cov_matrix}")
```

2. Eigen Decomposition and Identifying Principal Components:
*   Compute Eigen Values and Eigen Vectors.
*   Sort the eigenvalues in descending order and choose the top k eigenvectors corresponding to the highest eigenvalues.
*   Identify the Principal Components with the help of cumulative Sum plot.

```python
# eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eigh(cov_matrix)

# sort in descending order because we need higher eigenvalues on top
sorted_indices = np.argsort(eigenvalues)[::-1] # higher the eigenvalue, more important features it captures

# sort the eigenvalues in descending order
eigenvalues = eigenvalues[sorted_indices]

# rearrange the eigenvectors to match the sorted eigenvalues
eigenvectors = eigenvectors[:, sorted_indices]

# choose top k eigenvectors corresponding to the highest eigenvalues (principal components)
k = 50
top_k_eigenvectors = eigenvectors[:, :k]  # take the first k eigenvalues (each column is a principal component)
```

```python
# identify the principal components with the help of cumulative sum plot
explained_variance_ratio = eigenvalues / np.sum(eigenvalues) # calculate explained variance ratio
cumulative_explained_variance = np.cumsum(explained_variance_ratio) # calculate the cumulative sum of explained variance

plt.plot(cumulative_explained_variance)
plt.title("Cumulative Explained Variance")
plt.xlabel("Number of Principal Components")
plt.ylabel("Cumulative Explained Variance")
plt.grid(True)
plt.show()
```

3. Reconstruction and Experiment:
*   Reconstruction: Transform the original data by multiplying it with the selected eigenvectors(PCs) to obtain a lower-dimensional representation.
*   Experiments: Pick Four different combination of principal components with various ex- plained variance value and compare the result.
*   Display the Results and Evaluate.

### Reconstruction

```python
k = 50 # choose k principal components
components = eigenvectors[: , : k]
```

```python
compressed_data = np.dot(centered_data, components)
```

```python
# transform the original data by multiplying it with the selected eigenvectors (Principal Components)
decompressed_data = np.dot(compressed_data, components.T) + mean # add mean while decompressing
```

```python
plt.figure(figsize=(12,6))

# Original Image
plt.subplot(1, 3, 1)
plt.imshow(image_array, cmap="gray")
plt.title("Original Image")
plt.axis("off")

# Compressed Representation
plt.subplot(1, 3, 2)
plt.imshow(compressed_data, cmap="gray", aspect="auto")
plt.title(f"Compressed Image (k={k})")
plt.axis("off")

# Decompressed Image
plt.subplot(1, 3, 3)
plt.imshow(decompressed_data, cmap="gray")
plt.title("Decompressed Image")
plt.axis("off")
```

### Experiments

```python
k1 = 10
k2 = 20
k3 = 50
k4 = 100
k5 = 150

components1 = eigenvectors [:,: k1]
components2 = eigenvectors [:,: k2]
components3 = eigenvectors [:,: k3]
components4 = eigenvectors [:,: k4]
components5 = eigenvectors [:,: k5]
```

```python
compressed_data1 = np.dot(centered_data, components1)
compressed_data2 = np.dot(centered_data, components2)
compressed_data3 = np.dot(centered_data, components3)
compressed_data4 = np.dot(centered_data, components4)
compressed_data5 = np.dot(centered_data, components5)
```

```python
decompressed_data1 = np.dot(compressed_data1, components1.T) + mean
decompressed_data2 = np.dot(compressed_data2, components2.T) + mean
decompressed_data3 = np.dot(compressed_data3, components3.T) + mean
decompressed_data4 = np.dot(compressed_data4, components4.T) + mean
decompressed_data5= np.dot(compressed_data5, components5.T) + mean
```

### Displaying Results

```python
# Decompressed Images
plt.subplot(2,3,1)
plt.imshow(decompressed_data1, cmap='gray')
plt.title("10 Components")
plt.axis('off')

plt.subplot(2,3,2)
plt.imshow(decompressed_data2, cmap='gray')
plt.title("20 Components")
plt.axis('off')

plt.subplot(2,3,3)
plt.imshow(decompressed_data3, cmap='gray')
plt.title("50 Components")
plt.axis('off')

plt.subplot(2,3,4)
plt.imshow(decompressed_data4, cmap='gray')
plt.title("100 Components")
plt.axis('off')

plt.subplot(2,3,5)
plt.imshow(decompressed_data5, cmap='gray')
plt.title("150 Components")
plt.axis('off')
```

### Evaluation

We started by calculating the principal components, sorting the eigenvectors based on their corresponding eigenvalues in descending order. The principal components carry the information about the most variance in the image. So, the top eigenvectors represent the most significant features of the image, as higher eigenvalues capture more variances. Then, we determined the optimal number of principal components for reconstructing the image by analyzing the cumulative explained variance plot.

By experimenting with different numbers of components for reconstructing the image, we observed the following:
- 10 principal components: The image retains only a rough outline, capturing the most basic structure.
- 20 principal components: The image appears blurry, but important details and objects start becoming recognizable.
- 50 principal components: TThe image is almost same as the original and preserves most of the important information.
- 100 principal components: The reconstructed image is nearly identical to the original image.
- 150 principal components: Increasing the number of components further does not show any significant improvement in the quality of the reconstructed image.

This demonstrates how Principal Component Analysis (PCA) reduces dimensionality in image while keeping most of the essential details. In the case of above image, important information lie under 50, rest are just noises in the images. So, using about 50 to 100 components is sufficient to reconstruct the image without any noticeable loss in quality.
